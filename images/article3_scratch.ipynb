```python
%matplotlib inline
import matplotlib.pylab as plt
from matplotlib import rc, rcParams
import pandas as pd
import plotly
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.gridspec as gridspec


# Make use of TeX
rc('text',usetex=True)
# Change all fonts to 'Computer Modern'
rc('font',**{'family':'serif','serif':['Computer Modern']})

#matplotlib.rcParams['font.sans-serif'].insert(0, 'Liberation Sans')
#matplotlib.rcParams['font.sans-serif'].insert(0, 'Arial')
#matplotlib.rcParams['font.family'] = 'sans-serif'
#sns.set(font="Verdana")

return_type='dict'
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('text', usetex=True)
```

```python
df = pd.read_csv('woodsupply_data.csv', index_col='Year')
df['Softwood'] = df['Softwood Harvest'] / df['Softwood AAC']
df['Hardwood'] = df['Hardwood Harvest'] / df['Hardwood AAC']
```

```python
def fit_line(x, y):
    X = sm.add_constant(x, prepend=True) #Add a column of ones to allow the calculation of the intercept
    ols_test = sm.OLS(y, X,missing='drop').fit()
    """Return slope, intercept of best fit line."""
    X = sm.add_constant(x)
    return ols_test
```

```python
x = np.arange(1990, 2013)
ols_test = fit_line(x, df['Hardwood'])
ols_test.fittedvalues
#ols_test.summary()
```

```python
plt.figure(figsize=(6, 5))
gs = gridspec.GridSpec(2, 2, width_ratios=[5,1])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])
ax3 = plt.subplot(gs[2])
ax4 = plt.subplot(gs[3])

x = np.arange(1990, 2013)
plt.ylim([0, 1])
plt.xlim([1989, 2013])

ax1.scatter(x=x, y=df['Softwood'], facecolors='white', edgecolors='black')
ax1.plot(x, fit_line(x, df['Softwood']).fittedvalues, color='black')
ax1.set_ylabel('Proportion of AAC Harvested')

ax3.scatter(x=x, y=df['Hardwood'], facecolors='white', edgecolors='black')
ax3.plot(x, fit_line(x, df['Hardwood']).fittedvalues, color='black')
ax3.set_ylabel('Proportion of AAC Harvested')

sns.boxplot(df[['Softwood']], ax=ax2, color='white')
sns.boxplot(df[['Hardwood']], ax=ax4, color='white')

plt.setp(ax1.get_xticklabels(), visible=False)
plt.setp(ax2.get_yticklabels(), visible=False)
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax4.get_xticklabels(), visible=False)
plt.setp(ax4.get_yticklabels(), visible=False)

ax1.set_ylim([0,1])
ax2.set_ylim([0,1])
ax3.set_ylim([0,1])
ax4.set_ylim([0,1])

ax2.yaxis.set_label_position("right")
ax4.yaxis.set_label_position("right")
ax2.set_ylabel('Softwood')
ax4.set_ylabel('Hardwood')
ax2.yaxis.get_label().set_rotation(270)
ax4.yaxis.get_label().set_rotation(270)
ax1.set_xlim([1989, 2013])
ax3.set_xlim([1989, 2013])

plt.tight_layout()
plt.savefig('aac_consumption.pdf', format='PDF', bbox_inches='tight')
```
